```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification", model="joeddav/xlm-roberta-large-xnli")
text = "L'IA est en train de transformer de nombreuses industries"
candidate_labels = ["teknologi", "éducation", "économie", "divertissement", "environnement"]

# Menggunakan template hipotesis multi-bahasa
result = classifier(text, candidate_labels=candidate_labels, hypothesis_template="Ce texte est au sujet de {}.")
print("Zero-shot classification multi-bahasa result:", result)
```

```python
from transformers import pipeline

# Menggunakan model facebook/bart-large-mnli
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
text = "Faiz mencatat bahwa perkembangan teknologi AI memberikan dampak signifikan pada ekonomi global, mempengaruhi banyak sektor industri."
# Label kandidat
candidate_labels = ["technology", "education", "economy", "entertainment", "environment"]
result = classifier(text, candidate_labels=candidate_labels, hypothesis_template="This text is about {}.")
print("Zero-shot classification result:", result)
```

```python
from transformers import pipeline

# Menggunakan model zero-shot classification
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
text = " Pak prabowo berhasil terpilih menjadi presiden RI ke-8 setelah mengalahkan pesaingnya anies baswedan dan ganjar pranowo."
# Label kandidat
candidate_labels = ["technology", "education", "economy", "politics", "environment"]
result = classifier(text, candidate_labels=candidate_labels, hypothesis_template="This text is about {}.")

top_label = {"label": result["labels"][0], "score": result["scores"][0]}
print("Zero-shot classification top result:", top_label)
```

```python
from transformers import pipeline

# Menggunakan model default untuk text generation
generator = pipeline("text-generation")
result1 = generator("The future of artificial intelligence is")

print("Generated Text 1:", result1[0]['generated_text'])
```

```python
from transformers import pipeline

# Menggunakan model gpt2
generator = pipeline("text-generation", model="gpt2")
result2 = generator(
    "The future of artificial intelligence is",
    max_length=50,
    num_return_sequences=3,
    temperature=0.7,
    top_p=0.9
)

for idx, text in enumerate(result2, 1):
    print(f"Generated Text 2-{idx}:", text['generated_text'])
```

```python
from transformers import pipeline

# Membuat pipeline
unmasker = pipeline("fill-mask")

# Kalimat yang akan di isi fill-mask
masked_sentence = "Regular exercise like cycling and <mask> can improve heart health."
result = unmasker(masked_sentence, top_k=2)

# Menampilkan hasilnya
for idx, prediction in enumerate(result, 1):
    print(f"Prediction {idx}: {prediction['token_str']} with score: {prediction['score']:.4f}")
```

```python
from transformers import pipeline

# Membuat pipeline untuk menjawab pertanyaan
qa_pipeline = pipeline("question-answering")

# Definisi konteks dan pertanyaan
context = "Salah satu cabang olahraga populer di Indonesia adalah renang. Bagi orang yang tinggal dipedesaan olahraga ini dapat dilakukan di sungai, danau, atau pinggiran laut. Sedangkan di perkotaan olahraga ini biasa dilakukan di kolam renang yang dibuat khusus."
question = "renang adalah?"

#jawaban
result = qa_pipeline(question=question, context=context)
print("Jawaban:", result['answer'])
```

```python
from transformers import pipeline

# Membuat pipeline untuk menjawab pertanyaan
qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")

# Konteks
context = (
    "Manchester United telah memenangkan gelar juara Liga Primer Inggris (EPL) sebanyak 13 kali, yaitu pada "
    "tahun-tahun: 1992-1993, 1993-1994, 1995-1996, 1996-1997, 1998-1999, 1999-2000, 2000-2001, "
    "2002-2003, 2006-2007, 2007-2008, 2008-2009, 2010-2011, dan 2012-2013."
)
question = "Berapa kali Manchester United memenangkan gelar juara Liga Primer Inggris?"

# jawaban
result = qa_pipeline(question=question, context=context)
print(f"Jawaban: {result['answer']}")
```

```python
from transformers import pipeline

# Membuat pipeline
classifier = pipeline("sentiment-analysis")

# kalimat yang diuji
sentences = [
   "Mr. Dika bought a new motorcycle for his child who succeeded in becoming the class champion.",
   "Rafa graduated from college, becoming the best graduate at his campus.",
   "Mr. Husein lost his money when he was about to buy groceries at the store."

]

# Analisis kode
for idx, sentence in enumerate(sentences, 1):
    result = classifier(sentence)[0]  # Mengambil hasil pertama dari daftar
    print(f"Sentence {idx}: '{sentence}'")
    print(f"Sentiment: {result['label']} with score: {result['score']:.4f}")
    print("-" * 40)
```

```python
from transformers import pipeline

# Membuat pipeline penerjemahan bahasa indo to inggris
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-id-en")

# Teks yang akan diterjemahkan
texts_to_translate = [
    "Perkenalkan, nama saya faiz.",
    "Makanan favorit saya adalah ikan nila bakar.",
    "Hobi saya adalah berenang, dan saya juga suka memasak."
]

# Menerjemahkan setiap kalimat
results = translator(texts_to_translate)

# Menampilkan hasil penerjemahan
for idx, result in enumerate(results, start=1):
    print(f"Kalimat {idx}: {texts_to_translate[idx - 1]} -> Terjemahan: {result['translation_text']}")
```
